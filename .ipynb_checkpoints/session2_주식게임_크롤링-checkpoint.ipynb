# 주식 투자 게임

```python
class User :       # 클래스 선언시, 클래스 네임에는 꼭 앞글자는 대문자여야!   
    def __init__(self, name, money, stock):       # 붕어빵 틀 입장에서 붕어빵을 부르는 것 ; self
        self.name = name
        self.money = money 
        self.stock = stock 
```

```python
kim = User('kim', 10000, 10000)
```

```python
kim.name
```

```python
class User :       # 클래스 선언시, 클래스 네임에는 꼭 앞글자는 대문자여야!    
    
    team = 'tiger'
    
    def __init__(self, name, money, stock):       # 붕어빵 틀 입장에서 붕어빵을 부르는 것 ; self
        self.name = name
        self.money = money 
        self.stock = stock 
        
    def info(self):   # 모든 메소드(def)는 self 인자를 가져야 
        print(f"name is {self.name}")     # {} 안은 그냥 코드처럼 행동 / self.name = 가져온 붕어빵의 이름을 가져와
        print(f"name is {self.money}") 
        print(f"name is {self.stock}") 
```

```python
kim = User('kim', 10000, 10000)
```

```python
kim.info()
```

```python
kim.team # 붕어빵틀에서 나오는 붕어빵이 가지는 team = tiger
```

```python
class Stock : 
    def __init__(self, name, price, change): 
        self.name = name 
        self.price = price   
        self.change = change
        
    def info (self):
        print(f"name is {self.name}")     # {} 안은 그냥 코드처럼 행동 / self.name = 가져온 붕어빵의 이름을 가져와
        print(f"name is {self.price}") 
        print(f"name is {self.change}")
```

```python
user_name = input('이름을 입력하세요 : ') 
user_money = int(input ('투자금을 입력하세요 : ')) # input은 무조건 문자열을 반환하기 때문에 해당 투자금 입력값을 int값으로 가져오기 위해서는 앞에 int를 붙여주어야 함. 
```

```python
kim = User(user_name, user_money, 0)
```

```python
kim.info()
```

```python
wapple = Stock('wapple', 1000, 0)
```

```python
wapple.info()
```

```python
while True :    # 뒤에 있는 조건이 참이면 계속 반복 
    inp = input('?  :  ')  
    if inp == "종료" : 
        print("게임이 종료되었습니다.")
        break    # 게임 종료시 해당 while문 빠져나옴. 
    elif inp == "매입":
        print("매입이 완료되었습니다.")
    elif inp == "매도":
        print("매도가 완료되었습니다.")
    else :   # 위에서 제시한 if문 / elif문 외 다른 조건의 경우가 들어온다면 모두 else 아래에 있는 경우로 실행 
        print("잘못된 명령입니다.")
     # while True문 작성하면 바로 break문도 만들어주어야 ! 
```

# 크롤링

```python
import urllib.request as req     
from bs4 import BeautifulSoup
```

```python
url = "https://namu.wiki/w/%EC%A1%B0%EC%9C%A0%EB%A6%AC" 
###
header = {'User-Agent' : 'Mozilla/5.0'} 
response = req.Request(url, headers = header)
###
html = req.urlopen(response)   
# 일반적인 response는 파이썬이 읽을 수 없기 때문에 beautifulsoup로 pasing을 해서 읽어와야함 
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.find_all(class_="_2RBnxNOu")
```

```python
soup.find_all(class_="_2RBnxNOu", attrs={'alt':'조유리 Op.22 프로필', 'data-filesize':'47162'})[0]
```
